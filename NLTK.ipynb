```python
# we will grab a web page content then we will analyze the text to see what the page is about
import urllib.request

response = urllib.request.urlopen('http://www.netvueitsolutions.com/')

html = response.read()

print (html)
```

```python
# above content have html tags,we need to remove and get plain text
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,"html5lib")
text = soup.get_text(strip=True)

print (text)
```

```python
# Now we have a clean text from the crawled web page.
# let’s convert that text into tokens by splitting the text like this:
tokens = [t for t in text.split()]

print (tokens)
```

```python
# Count word frequency
freq = nltk.FreqDist(tokens)

for key,val in freq.items():

    print (str(key) + ':' + str(val))
```

```python
#  the most frequent token is IT.we can plot a graph for those tokens using plot function like this:
freq.plot(20, cumulative=False)
```

```python
# Remove stop words using NLTK,To get English stop words, you can use this code:
from nltk.corpus import stopwords

stopwords.words('english')
```

```python
#  Remove stop words using NLTK
clean_tokens = tokens[:]

sr = stopwords.words('english')

for token in tokens:

    if token in stopwords.words('english'):

        clean_tokens.remove(token)
```

```python
freq = nltk.FreqDist(clean_tokens)

for key,val in freq.items():

    print (str(key) + ':' + str(val))
```

```python
freq.plot(20,cumulative=False)
```
